### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Create the figure and 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line Chart (Top-left)
x = np.arange(0, 11, 1)
y = x**2
axes[0, 0].plot(x, y, marker='o', linestyle='-')
axes[0, 0].set_title('Line Plot')
axes[0, 0].set_xlabel('X Values')
axes[0, 0].set_ylabel('Y Values')
axes[0, 0].grid(True)

# Bar Chart (Top-right)
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']
axes[0, 1].bar(categories, values, color=colors)
axes[0, 1].set_title('Bar Chart')

# Scatter Plot (Bottom-left)
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axes[1, 0].scatter(x_scatter, y_scatter, color='blue')
for i in range(len(x_scatter)):
    axes[1, 0].annotate(f'({x_scatter[i]}, {y_scatter[i]})', 
                         (x_scatter[i], y_scatter[i]), 
                         textcoords="offset points", xytext=(5,5), ha='center')
axes[1, 0].set_title('Scatter Plot')

# Pie Chart (Bottom-right)
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'])
axes[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
